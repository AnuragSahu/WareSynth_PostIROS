```python
import numpy as np
import os
```

```python
def writeAnn(BoxAnnArr, RackAnnArr, fol):
    with open("./" +"ann_" + fol + ".txt", "+a") as f:
        f.write("Box,")
        f.write("\nBox,".join(",".join(map(str, x)) for x in BoxAnnArr))
        f.write("\nShelf,")
        f.write("\nShelf,".join(",".join(map(str, x)) for x in RackAnnArr))
        f.write("\n")
    return 

def read_annFile(filepath):
    with open(filepath) as f:
        lines = [line.rstrip() for line in f]
    ann0 = []
    ann0_boxes = []
    ann0_racks = []
    for ll in lines:
        line = ll.split(", ")
        a = line[:]
        a = np.array(a)
        if a[0] == "Box":
            ann0_boxes.append(a[1:].astype(np.float))
        else:
            ann0_racks.append(a[1:].astype(np.float))
        ann0.append(a[:])
    ann0 = np.array(ann0)
    ann0_boxes = np.array(ann0_boxes)
    ann0_racks = np.array(ann0_racks)
    return ann0, ann0_boxes, ann0_racks

def findCorrespondances(box0, box1):
    margin_yz = 2
    shape0 = box0.shape
    shape1 = box1.shape
    p0, p1 = 0, 0
    cor0 = []
    cor1 = []
    corres_indices = []

    while p0 < shape0[0] and p1 < shape1[0]:
#         print(box0[p0, -3])
        if p0 == 0 and box0[p0, -3] < 4:
#             print("Box vanishing!!")
            corres_indices.append((p0, -1))
            p0+=1
        elif np.linalg.norm(box0[p0, -2:] - box1[p1, -2:]) < margin_yz:
            cor0.append(box0[p0])
            cor1.append(box1[p1])
            corres_indices.append((p0, p1))
            p0+=1 
            p1+=1     
        elif np.linalg.norm(box0[p0, -2:] - box1[p1 + 1, -2:]) < margin_yz:
            cor0.append(box0[p0])
            cor1.append(box1[p1 + 1])
#             print(p0, p1)
#             if corres_indices[-1][0] != p0 - 1:
#                 corres_indices.append((p0, -1))
            corres_indices.append((p0, p1+1))
            p0+=1 
            p1+=2
        elif np.linalg.norm(box0[p0 + 1, -2:] - box1[p1, -2:]) < margin_yz:    
            cor0.append(box0[p0 + 1])
            cor1.append(box1[p1])
            corres_indices.append((p0+1, p1))
            p0+=2 
            p1+=1
    if p1 <= shape1[0] - 1:
        while p1 < shape1[0]:
            corres_indices.append((-1, p1))
            p1+=1
    return np.array(cor0), np.array(cor1), corres_indices

# Pass correspondances
def findTranlation(cor0, cor1):
    n0, n1 = cor0.shape
    margin_x = 2
    corr1 = []
    corr0 = []
    avg_translation = np.array([0., 0., 0.])
    pt = 0
    for i in range(n0):
        if np.linalg.norm(cor0[i, 3] - cor1[i, 3]) < margin_x:
            avg_translation += (cor0[i, :3] - cor1[i, :3])
            pt+=1
    avg_translation = avg_translation / pt
    return avg_translation

def findMergedDis(frames_per_frame, curr_row, curr_col, path_to_ann, fol):
    filepaths = [x for x in os.listdir(path_to_ann) if x.endswith(".txt") and x.startswith(f"ann_{curr_row}_{curr_col}")]
    filepaths.sort()
#     print("\n", filepaths)
    finalBoxAnn = []
    finalRackAnn = []
    ret = 0
    for filenum in range(len(filepaths) - 1):
        filepath_0 = filepaths[filenum]
        filepath_1 = filepaths[filenum + 1]
        ann_0, boxes_0, racks_0 = read_annFile(path_to_ann + filepath_0)
        ann_1, boxes_1, racks_1 = read_annFile(path_to_ann + filepath_1)

        corBox_0, corBox_1, corBox_indices = findCorrespondances(boxes_0, boxes_1)

        avgDim_transBox = np.array([4])
        rack_space = 10
        rack_len = 162
        
        if filenum%frames_per_frame == 0:
            for nn in range(boxes_0.shape[0]):
                finalBoxAnn.append(boxes_0[nn])
            for nn in range(racks_0.shape[0]):
                finalRackAnn.append(racks_0[nn])  
       
        if (corBox_indices[-1][0] == corBox_indices[-1][1]) or (boxes_0.shape[0] - boxes_1.shape[0]) == (corBox_indices[-1][0] - corBox_indices[-1][1]):
#             print("In first IF!!!")
            if abs(racks_0[-1][3]  -  racks_1[-1][3]) < 5*avgDim_transBox[0]:
#                 print("In if 1!")
                prev = finalRackAnn[-1][3]
                new = racks_1[-1][3]
                finalRackAnn[-1][3] = racks_1[-1][3]
                finalRackAnn[-1][0] += (new - prev)/2
            else:
#                 print("In else 1!")
                racks_11 = racks_1.copy()
                racks_1[-1][0] = finalRackAnn[-1][0] + finalRackAnn[-1][3]/2 + rack_space + racks_1[-1][3]/2 
                finalRackAnn.append(racks_1[-1])
                racks_1 = racks_1.copy()
                
            if abs(boxes_0[-1][3] - boxes_1[-1][3]) > 0.85*avgDim_transBox[0]:
                prev = finalBoxAnn[-1][3]
                new = boxes_1[-1][3]
                if prev > new:
                    boxes_11 = boxes_1.copy()
                    boxes_1[-1][0] = (finalRackAnn[-1][0] - finalRackAnn[-1][3]/2) + (boxes_1[-1][0] - boxes_1[-1][3]/2 - (racks_1[-1][0]  - racks_1[-1][3]/2))
                    finalBoxAnn[-1] = boxes_1[-1]
                    boxes_1 = boxes_11.copy()
                    
                else:
                    finalBoxAnn[-1][3] = boxes_1[-1][3]
                    finalBoxAnn[-1][0] += (new - prev)/2 
        
        else:
#             print("In first ELSE!!!")
#             print(abs(finalRackAnn[-1][3]  -  racks_1[-1][3]), finalRackAnn[-1][3], racks_1[-1][3])
            if abs(finalRackAnn[-1][3]  -  racks_1[-1][3]) < 5*avgDim_transBox[0]:
#                 print("In if 2!")
                prev = finalRackAnn[-1][3]
                new = racks_1[-1][3]
                finalRackAnn[-1][3] = racks_1[-1][3]
                finalRackAnn[-1][0] += (new - prev)/2
            else:
#                 print("In else 2!")
                racks_11 = racks_1.copy()
                racks_1[-1][0] = finalRackAnn[-1][0] + finalRackAnn[-1][3]/2 + rack_space + racks_1[-1][3]/2 
                
                finalRackAnn.append(racks_1[-1])
                racks_1 = racks_11.copy()
            
            boxes_11 = boxes_1
            boxes_11[-1][0] = (finalRackAnn[-1][0] - finalRackAnn[-1][3]/2) + (boxes_1[-1][0] - boxes_1[-1][3]/2 - (racks_1[-1][0]  - racks_1[-1][3]/2)) + boxes_1[-1][3]/2
            finalBoxAnn.append(boxes_11[-1])
#         if fol == "middle_ann" and filenum == 29 and curr_row == 2 and curr_col == 1 :
#             print("_______zsdfghjkl__________")
#             ret = 1
#             break
    finalBoxAnn = np.array(finalBoxAnn)
    finalRackAnn = np.array(finalRackAnn)
    writeAnn(finalBoxAnn, finalRackAnn, fol)
    return finalBoxAnn, finalRackAnn, ret
```

```python
folders = ["bottom_ann", "middle_ann", "top_ann"]
num_rows = 2
num_cols = 2
frames_per_frame = 70
os.remove("ann_bottom_ann.txt")
os.remove("ann_middle_ann.txt")
os.remove("ann_top_ann.txt")
for folder in folders:
    for row in range(num_rows):
        for col in range(num_cols):
            finalBoxAnn, finalRackAnn, ret = findMergedDis(frames_per_frame, row + 1, col + 1, "./" + folder + "/", folder)
```
